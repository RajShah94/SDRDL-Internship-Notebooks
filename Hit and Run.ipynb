# Objective:

1. Find all the cases between 2009 and 2012 for which the final sentence (highest_charge) is PC187, PC191, PC192, and VC20001

2. Create a yearly summary for each that includes:
	- How many were incarcerated and for how long.
	- How many were not incarcerated.

3. Infer reasonable values for the 2008 year, by averaging the 2009-2012 summary.

About final disposition: Please exclude dismissed cases, and blanks (NAs) as these cases are still in development.


The objective is to assist Mr. Koran in his investigation:

“This way I'd write something like: out of 6,900 hit and run accidents (that resulted in injury or death) between 2008 and 2012, fewer than X in 10 were caught and punished. This keeps the focus on the punishment, or consequences.

From there, we can say: Out of those who were punished, only X many did any time behind bars, etc."

So now we create a subset in which the highest disposed charge is either PC187, PC191, PC192 or VC20001

```python
hdcDF = df[(df.highest_dispo_charge == 'PC187(a)') |
        (df.highest_dispo_charge == 'VC20001(a)') | 
        (df.highest_dispo_charge == 'VC20001(b)(2)') | 
        (df.highest_dispo_charge == 'PC191.5(a)') |
        (df.highest_dispo_charge == 'PC192(c)(1)') | 
        (df.highest_dispo_charge == 'PC191.5(b)') | 
        (df.highest_dispo_charge == 'PC187(a) - PC664') |
        (df.highest_dispo_charge == 'PC192(c)(1), unlawful act') |
        (df.highest_dispo_charge == 'PC192(c)(1), lawful act') |
        (df.highest_dispo_charge == 'VC20001(a) - PC664')] 
```

Similarly, we create a subset  in which the highest charge is either PC187, PC191, PC192 or VC20001

```python
hcDF = df[(df.highest_charge == 'PC187(a)') |
        (df.highest_charge == 'VC20001(a)') | 
        (df.highest_charge == 'VC20001(b)(2)') | 
        (df.highest_charge == 'PC191.5(a)') |
        (df.highest_charge == 'PC192(c)(1)') | 
        (df.highest_charge == 'PC191.5(b)') | 
        (df.highest_charge == 'PC187(a) - PC664')] 
```

Once we know that our highest charge subset has dismissed and blank cases, we exclude them.

```python
hcDF = hcDF[(hcDF.final_disposition != nan) &
            (hcDF.final_disposition != 'Dismissed (IFJ / IOE)') & 
            (hcDF.final_disposition != 'Dismissed Other (see note)') &
            (hcDF.final_disposition != 'Dismissed by Court (PC 1118.1)') &
            (hcDF.final_disposition != 'Dismissed - Defendant Deceased')]
```
