```python
df = pd.read_csv('sweetwater_transfers.csv')
df
```

```python
applied = df.groupby(['year','dest']).id.count()
accepted = df[df.status == "Approved"].groupby(['year','dest']).id.count()
rejected = df[df.status != "Approved"].groupby(['year','dest']).id.count()
released = df[df.status == "Approved"].groupby(['year','origin']).id.count()

flow = pd.DataFrame({'Applied':pd.Series(applied),
                     'Accepted':pd.Series(accepted),
                     'Denied':pd.Series(rejected),
                     'Released':pd.Series(released)})
flow = flow.fillna(0)
net = flow['Accepted'] - flow['Released']
flow['Net'] = pd.Series(net)
flow = flow[['Applied','Accepted','Denied','Released','Net']]
flow
```

```python
applied = df.groupby('dest').id.count()
accepted = df[df.status == "Approved"].groupby('dest').id.count()
rejected = df[df.status != "Approved"].groupby('dest').id.count()
released = df[df.status == "Approved"].groupby('origin').id.count()

flow1 = pd.DataFrame({'Applied':pd.Series(applied),
                     'Accepted':pd.Series(accepted),
                     'Denied':pd.Series(rejected),
                     'Released':pd.Series(released)})
flow1 = flow1.fillna(0)
net = flow1['Accepted'] - flow1['Released']
flow1['Net'] = pd.Series(net)
denial_rate = ((flow1['Applied'] - flow1['Accepted'])/flow1['Applied'])*100
flow1['% Denied'] = pd.Series(denial_rate)
flow1 = flow1.fillna('Does not apply')
flow1 = flow1[['Applied','Accepted','Denied','% Denied','Released','Net']]
flow1
```
